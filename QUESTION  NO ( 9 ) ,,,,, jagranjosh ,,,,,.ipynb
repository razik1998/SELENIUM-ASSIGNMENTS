```python
#   Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead, 
#        Term of office, Remarks) from https://www.jagranjosh.com/.
```

```python
#  IMPORTING REQUIRED LABRARIES

import selenium 
import pandas as pd 
from selenium import webdriver
import warnings
warnings.filterwarnings ('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
#  CONNECTING TO THE DRIVER
driver=webdriver.Chrome(r"C:\Users\ASUS\Downloads\chromedriver.exe")
```

```python
#  OPENING THE JAGRANJOSH PAGE ON AUTOMATED CHROME BROWSER
driver.get("https://www.jagranjosh.com/.")
```

```python
#   ENTERING DESIGNATION AS REQUIRED 
designation = driver.find_element (By.XPATH,"/html/body/div[1]/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a")
designation.send_keys ('GK')
```

```python
#  CLICKING SEARCH BUTTON 
search = driver.find_element (By.XPATH,"/html/body/div[1]/div/div/div[2]/div/div[10]/div/div/ul/li[2]/a")
search.click()
```

```python
#  CREATING EMPTY LIST
NAMES =[]
BORN_DEAD =[]
TERM_OFFICE =[]
REMARK = []
```

```python
# SCRAPING NAMES 
prime = driver.find_elements (By.XPATH,"//strong")
for i in prime[0:18] :
    name = (i.text)
    NAMES.append (name)
    
    
    
# SCRAPING BORD DEAD 
born = driver.find_elements (By.XPATH,"//p")
for i in born [0:18]:
    dead = (i.text)
    BORN_DEAD.append (dead)
    
    
    
# SCRAPING TERMS OF OFFICE
term = driver.find_elements (By.XPATH,"//p")
for i in term [0:18]:
    office = (i.text)
    TERM_OFFICE.append (office)
    
    
    
# SCRAPING REMARKS
remarks = driver.find_elements (By.XPATH,"//p")
for i in remarks [0:18]:
    mark = (i.text)
    REMARK.append (mark)
    
    
    
```

```python
# printing lenghts of elements
print(len(NAMES),len(BORN_DEAD),len(TERM_OFFICE),len(REMARK))
```

```python
# MAKING DATA FRAME 
df = pd.DataFrame ({'NAMES':NAMES,'BORN_DEAD':BORN_DEAD,'TERM_OFFICE':TERM_OFFICE,'REMARK':REMARK})
df
```
