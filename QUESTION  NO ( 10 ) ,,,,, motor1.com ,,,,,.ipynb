```python
#  IMPORTING REQUIRED LABRARIES

import selenium 
import pandas as pd 
from selenium import webdriver
import warnings
warnings.filterwarnings ('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
#  CONNECTING TO THE DRIVER
driver=webdriver.Chrome(r"C:\Users\ASUS\Downloads\chromedriver.exe")
```

```python
#  OPENING THE NAUKRI PAGE ON AUTOMATED CHROME BROWSER
driver.get("https://www.motor1.com/")
```

```python
#  CLICKING SEARCH BUTTON 
search = driver.find_element (By.XPATH,"/html/body/div[4]/div[1]/div[3]/ul/li[6]/ul/li[1]")
search.click()
```

```python
#  CLICKING SEARCH BUTTON 
search = driver.find_element (By.XPATH,"/html/body/div[3]/div[8]/div[1]/div[1]/div/div/div[8]/div/div[1]/h3")
search.click()
```

```python
# CREATING EMPTY LIST
CAR_NAMES = []
CAR_PRICES= []
```

```python
#  SCRAPING CAR NAMES IN ONE GO
crn = driver.find_elements (By.XPATH,'//h3[@class="subheader"]')
for i in crn [0:50]:
    name = i.text
    CAR_NAMES.append (name)
```

```python
# SCRAPING CARS PRICE IN ONE GO
car_prc = driver.find_elements (By.XPATH,"//strong")
for i in car_prc[0:50]:
    price = i.text
    CAR_PRICES.append (price)
    
```

```python
# PRINTING LENGHTS IN ONE GO
print (len(CAR_NAMES),len(CAR_PRICES))
```

```python
# MAKING DATA FRAME AND DISPLAY IN DATA FRAME FORMAT
df = pd.DataFrame ({'CAR_NAMES':CAR_NAMES,'CAR_PRICES':CAR_PRICES})
df
```
